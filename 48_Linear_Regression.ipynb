```python
# Linear Regression - Linear Regression is a supervised learning algorithm that predicts a continuous target
#                     by fitting a linear relationship between features and the outcome using least squares.

# It is of three types -
# 1. Simple LR - one i/p (predicting 'get placement or not' by 'CGPA' )
# 2. Multiple LR - more than one i/p (predicting 'get placed or not' by 'CGPA' and 'College')
# 3. Polynomial LR - when the data is not linear dependent

# Terminology - sorcostic error, best-fit line
```

```python
import pandas as pd
```

```python
df = pd.read_csv('placement.csv')
df.head(5)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(x = df['cgpa'], y = df['package'])
plt.xlabel('CGPA')
plt.ylabel('Package in lakhs')
```

```python
from sklearn.model_selection import train_test_split
X = df.iloc[:,0:1]
y = df.iloc[:,-1]
X_train, X_test,y_train,y_test = train_test_split(X,y, test_size=0.2 , random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
X_test
```

```python
lr.predict(X_test.iloc[0].values.reshape(1,1)) # we are accessing dataframe which is 2D and want one value so
                                               # we have to reshape this into (1,1) and we want values of it
                                               # that's why we have used it values
```

```python
plt.scatter(x = df['cgpa'], y = df['package'])
plt.plot(X_train, lr.predict(X_train), color = 'red')
plt.xlabel('CGPA')
plt.ylabel('Package in lakhs')
```

```python
# Note- now we have line, it means we also have slope of line and intercept of line also
m = lr.coef_ # this will give us line
b = lr.intercept_ # this will give intercept
```

```python
# Equation of line => y = mx + b , m can refer as weight it means y depends mostly on m 

m * 8.58 + b # this is exactly same which we have calculated
```
