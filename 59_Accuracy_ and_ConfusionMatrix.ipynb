```python
# Accuracy = No. of correct / total no. of Predictions
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('heart.csv')
df.sample(7)
```

```python
# train - test split
X = df.drop(columns=['target'])
y = df['target']
X_train, X_test, y_train, y_test = train_test_split(X,y, random_state=42, test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
lgr = LogisticRegression()
lgr.fit(X_train, y_train)
y_pred1 = lgr.predict(X_test)
```

```python
dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)
y_pred2 = dt.predict(X_test)
```

```python
print("accuracy_score of Logistic Regression: ", accuracy_score(y_test, y_pred1))
print("accuracy_score of Decision tree: ", accuracy_score(y_test, y_pred2))
```

```python
from sklearn.metrics import confusion_matrix
cm1 = confusion_matrix(y_test, y_pred1)
print(cm1)
cm2 = pd.DataFrame(confusion_matrix(y_test,y_pred2), columns=[0,1])
print(cm2)
```
