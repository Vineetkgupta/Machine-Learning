```python
import pandas as pd
```

```python
df = pd.read_csv('titanic_train.csv')
df.head()
```

*Categorical Data*

```python
# 1. Countplot
import seaborn as sns
import matplotlib.pyplot as plt
# df['Survived'] = df['Survived'].astype(int)
sns.countplot(df['Sex'])
df['Survived'].value_counts()
```

```python
sns.countplot(df['Embarked'])
pd.value_counts(df['Embarked'])
# pd.value_counts(df['Survived']).plot(kind='bar') # other way of plotting bar graph
```

```python
# 2. Pie Chart
pd.value_counts(df['Survived']).plot(kind='pie', autopct = '%.2f')
```

2. Numerical Data

```python
# a. Histogram
plt.hist(df['Age'], bins = 8)
# Using sns
sns.histplot(df['Age'], bins = 8)
```

```python
# b. Displot
sns.distplot(df['Age'])
```

```python
# C. Boxplot
sns.boxplot(df['Age'])
```
