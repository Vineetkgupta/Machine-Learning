# 3. OneHotEncoding using sklearn

```python
# train - test split
X_train, X_test, y_test, y_train = train_test_split(X,y, random_state=0, test_size=0.2)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop='first')
```

```python
X_train_new = ohe.fit_transform(X_train[['fuel', 'owner']]).toarray()# here we get sparse matrix so to convert it into array
                                                                     # we are using toarray()
X_test_new = ohe.fit_transform(X_test[['fuel', 'owner']]).toarray()
```

# 4. OneHotEncoding with Top Categories

```python
counts = df['brand'].value_counts()
counts
```

```python
df['brand'].nunique()
threshold = 100
```

```python
repl = counts[counts <= threshold].index
```

```python
pd.get_dummies(df['brand'].replace(repl, 'uncommon')).astype(int) # here, we are using astype as int as new version of get_dummies return data in form of true and false
```
