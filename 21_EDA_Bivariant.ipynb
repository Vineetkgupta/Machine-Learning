```python
import pandas as pd
import seaborn as sns
```

```python
tips = pd.read_csv('tip.csv')
tips.head()
```

```python
titanic = pd.read_csv('titanic_train.csv')
titanic.head()
```

```python
flights = pd.read_csv('flights.csv')
flights.head()
```

```python
iris = pd.read_csv('Iris.csv')
iris.head()
```

# 1. Scatterplot => Numerical - Numerical

```python
sns.scatterplot(x= 	tips['total_bill'], y= tips['tip'])
```

```python
sns.scatterplot(x= 	tips['total_bill'], y= tips['tip'], hue=tips['sex']) # hue is optional it is used to add color on the basis 
                                                                         # of sex- Multivariate
```

```python
sns.scatterplot(x= 	tips['total_bill'], y= tips['tip'], hue=tips['sex'], style=tips['smoker']) # here we added one more info
```

```python
sns.scatterplot(x= 	tips['total_bill'], y= tips['tip'], hue=tips['sex'], style=tips['smoker'], size = tips['size'])
```

# 2. Bar Plot => Categorical + Numerical

```python
titanic.head()
```

```python
sns.barplot(x =titanic['Pclass'],y = titanic['Age'], hue=titanic['Sex']) 
```

# 3. Box Plot => Numerical - Category

```python
sns.boxplot(x=titanic['Sex'], y = titanic['Age'], hue = titanic['Survived'])
```

# 4. Distplot => Numerical - Category

```python
# In version 0.11 distplot is removed from the seaborn, instead of it we use histplot with kde = True
sns.histplot(titanic[titanic['Survived'] == 0]['Age'], kde = True)
```

```python
# If you want only kde then don't use histplot, use kdeplot
sns.kdeplot(titanic[titanic['Survived'] == 0]['Age']) # blue color
sns.kdeplot(titanic[titanic['Survived'] == 1]['Age']) # Orange Color
# This is the beauty of EDA as it shows what is the number of people of different age group survived vs not survived , it can 
# be seen clearly from the kde plot 
```

# 5. Heatmap => Categorical - Categorical

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
titanic.groupby('Pclass')['Survived'].mean()* 100 # This shows how many percentage of people survived from each class
```

```python
titanic.groupby('Embarked')['Survived'].mean() * 100
```

# 6. Cluster Map=> Categorical - Categorical

```python
sns.clustermap(pd.crosstab(titanic['SibSp'], titanic['Survived']))
pd.crosstab(titanic['SibSp'], titanic['Survived'])
```
