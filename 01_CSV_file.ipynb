**In this we will see how to work with csv files**

```python
import pandas as pd
```

```python
# Opening a local CSV file
df = pd.read_csv('aug_train.csv')
df
```

```python
# Sep Parameter
df = pd.read_csv('movie_titles_metadata.tsv', sep = '\t')
print(df) # here we can see it does not have column name so it treats first rows as column name. To overcome this we will 
          # give columns names accordingly
df = pd.read_csv('movie_titles_metadata.tsv', sep = '\t', names=['sno','Movie','Release Year','Rating','Budget','Genre'])
df
```

```python
# Index col parameter - if we want to create any column of our dataset as index
pd.read_csv('aug_train.csv', index_col='enrollee_id') 
```

```python
# Header parameter - if your dataset have column name but it behaves as a row then you can use header = 1
# pd.read_csv('test.csv', header = 1)
```

```python
# use_cols parameter - in dataset there are multiple columns but you want some of them then you select this
pd.read_csv('aug_train.csv', usecols=['enrollee_id', 'city','gender'])
```

```python
# Skiprows/nrows Parameter
pd.read_csv('aug_train.csv', skiprows=[0,1])
# pd.read_csv('aug_train.csv', skiprows=[0:3])
# pd.read_csv('aug_train.csv',nrows = 100)
```

```python
# Encoding Parameter - if it is not in UTF data format
pd.read_csv('BX-Books.csv', encoding='latin-1')
```

```python
# Skip bad lines
# pd.read_csv('BX-Books.csv', encoding='latin-1',error_bad_lines = False)
```

```python
# dtypes Parameter
pd.read_csv('aug_train.csv',dtype={'target':int}).info()
```

```python
# Handling Dates
pd.read_csv('IPL Matches 2008-2020.csv', parse_dates=['date']).info() # now we can implement all the operations related to data
```

```python
def rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    else: return name
```

```python
rename("Royal Challengers Bangalore")
```

```python
# Convertor
pd.read_csv('IPL Matches 2008-2020.csv', converters={'team1':rename})
```

```python
# na_values parameters - missing values
pd.read_csv('aug_train.csv', na_values=['Male'] )
```

```python
# Loading a huge dataset in chunks
dfs = pd.read_csv('aug_train.csv', chunksize=5000)
```
